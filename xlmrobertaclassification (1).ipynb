```python
from transformers import XLMRobertaForSequenceClassification, Trainer, TrainingArguments, AutoTokenizer
from datasets import load_dataset

# Load the dataset from Hugging Face datasets library
dataset = load_dataset('tamilmixsentiment')

# Preprocess the dataset
def preprocess_function(examples):
    return tokenizer(examples['text'], padding='max_length', truncation=True, max_length=128)

# Load the tokenizer associated with XLM-RoBERTa
tokenizer = AutoTokenizer.from_pretrained('xlm-roberta-base')

train_dataset = dataset['train'].map(preprocess_function)

# Load the pre-trained XLM-RoBERTa model
model = XLMRobertaForSequenceClassification.from_pretrained('xlm-roberta-base', num_labels=5)

# Define training arguments
training_args = TrainingArguments(
    output_dir='./output',  # Specify the output directory
    per_device_train_batch_size=8,
    num_train_epochs=3,
    learning_rate = 5e-6,
    weight_decay = 1e-3,
    logging_dir='./logs',
)

# Create Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
)

# Fine-tune the model
trainer.train()

# Save the trained model
model_path = "./fine_tuned_model"
trainer.save_model(model_path)
```

```python
from transformers import Trainer
from datasets import load_dataset
from sklearn.metrics import accuracy_score

# Define a function to compute accuracy
def compute_accuracy(pred):
    labels = pred.label_ids
    preds = pred.predictions.argmax(-1)
    return {"accuracy": accuracy_score(labels, preds)}

# Load the validation dataset
validation_dataset = load_dataset("tamilmixsentiment", split="validation")

# Preprocess the validation dataset
validation_dataset = validation_dataset.map(preprocess_function)

# Evaluate the trained model on the validation dataset
trainer = Trainer(model=model, compute_metrics=compute_accuracy)
eval_results = trainer.evaluate(eval_dataset=validation_dataset)

# Print accuracy
print("Accuracy:", eval_results["eval_accuracy"])
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification, pipeline

# Load the tokenizer and model
tokenizer = AutoTokenizer.from_pretrained('xlm-roberta-base')
model = AutoModelForSequenceClassification.from_pretrained('./fine_tuned_model')

# Define a function to predict sentiment
def predict_sentiment(sentence):
    # Tokenize the sentence
    inputs = tokenizer(sentence, padding='max_length', truncation=True, max_length=128, return_tensors="pt")

    # Perform inference
    outputs = model(**inputs)
    logits = outputs.logits
    probabilities = logits.softmax(dim=-1)

    # Get the predicted label
    predicted_label_id = probabilities.argmax().item()
    predicted_label = model.config.id2label[predicted_label_id]

    return predicted_label

# Example usage
new_sentence = "mass padam da"
predicted_sentiment = predict_sentiment(new_sentence)
print("Predicted sentiment:", predicted_sentiment)
```
